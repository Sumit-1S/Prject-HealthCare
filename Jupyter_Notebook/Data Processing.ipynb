```python
import pandas as pd
import math
```

```python
df = pd.read_csv('../Dataset/dataset.csv')
```

```python
df.head()
```

```python
data = {}
for i in range (0,len(df)):
    disease = df.iloc[i]['Disease']
    if disease not in data:
        data[disease] = {}
    for j in range (1,len(df.iloc[i])):
        symp = df.iloc[i][j]
        if isinstance(symp,float) and math.isnan(symp):
            break
        symp = symp.strip()
        if symp not in data[disease]:
            data[disease][symp] = 1
        else:
            data[disease][symp]+= 1

data
```

```python
df_symptoms = pd.read_csv('../Dataset/Symptom-severity.csv')
df_symptoms.head()
```

```python
new_data = {}
columns = []
for (key,val) in zip(data,data.values()):
    columns.append(key)
    for symp in df_symptoms['Symptom']:
        if symp not in new_data:
            new_data[symp] = []
        if symp in val:
            new_data[symp].append(val[symp])
        else:
            new_data[symp].append(0)
print(new_data)
print(columns)
```

```python
df = pd.DataFrame.from_dict(new_data, orient='index')
df.dropna(axis=1,inplace=True)
df.columns = columns
df['symptoms'] = df.index
df = df.reset_index()
df
```

```python
new_df = df
new_df
```

```python
new_df.to_csv('../Dataset/symptoms_for_analysis.csv')
```
